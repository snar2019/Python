##downward trend in MPG as the number of gears increases. But it's subtle.

```python
sns.lmplot(x= '# Gears', y='CombMPG', data=df)
```

##overall trend of lower MPG's with more gears is apparent. More gears isn't better when it comes to efficiency.

```python
sns.jointplot(x='# Gears', y= 'CombMPG', data=df)
```

##The most common gear configuration seems to be 8, with 6 closely behind. And MPG ratings seem to roughly follow a bell curve centered at around 22 or so.
 

```python
sns.set(rc={'figure.figsize' : (15,5)})
bx=sns.boxplot(x='# Gears', y='CombMPG', data=df)
bx.set_xticklabels(bx.get_xticklabels())
```

##. many of the extreme values are in fact outliers that are best discarded when analyzing the trends. This makes the real relationships easier to see; those continuously variable transmissions with a single gear are really quite good at fuel efficiency (higher MPG's are more efficient).Between 5 and 8 things are roughly the same, but from 8-10 things decline markedly.

```python
bx=sns.swarmplot(x='# Gears', y='CombMPG', data=df)
bx.set_xticklabels(bx.get_xticklabels())
```

##.The swarm plot makes it even more apparent that those high MPG outliers are in fact outliers on the 6-gear vehicles; the vast majority of 6-gear vehcles have less than 32 MPG. And the overall trend is perhaps easiest to visualize and understand in this representation.

So we can conclude that more gears result in worse fuel efficiency. Which strikes me as a little counter-intuitive! But, this is real data from the US Department of Energy for 2019 model year cars, and that's a real conclusion we've drawn, thanks to Seaborn
